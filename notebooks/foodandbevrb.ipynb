```python
import pandas as pd
from pathlib import Path

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pprint
import json
import numpy as np
import requests
import hvplot.pandas
```

```python
# Load retail and food sell data from files

non_store_df = Path("Nonstore_Retailers_Total.csv")
retailers = pd.read_csv(non_store_df)
print(retailers)
```

```python
food_drink_df = Path("Food _Serv_ Drink _Places_Total.csv")
food = pd.read_csv(food_drink_df)
print(food)
```

```python
covid_cases = Path("WHO-COVID-19-global-data.csv")
co_spread = pd.read_csv(covid_cases)
```

```python
co_spread.head
```

```python
co_us_data = co_spread[co_spread['Country']=='United States of America']
```

```python
co_us_data.head()
```

```python
co_us_data.sort_values("Date_reported")
```

```python
co_us_data.reset_index(inplace=True)
co_us_data.head()
```

```python
# Daily infections / daily deadths
x = co_us_data['index']-110500
y1 = co_us_data[['New_cases']]
y2 = co_us_data[['New_deaths']]
fig = plt.figure()
axes = fig.add_subplot(111)

axes.scatter(x, y1, s=2, c='b', marker="s", label='New daily infections')
axes.scatter(x, y2, s=2, c='r', marker="o", label='New daily deaths')
plt.ylabel("number of Infections")
plt.xlabel('daily new deaths')
plt.legend(loc='upper left');
plt.show()
plt.savefig('Matplotlib_save_plot.png')
```

```python
#  Daily retail sells / daily food sells
x = retailers.index
y1 = retailers[['Value']]
y2 = food[['Value']]
fig = plt.figure()
axes = fig.add_subplot(111)

axes.scatter(x, y1, s=2, c='b', marker="s", label='retailers')
axes.scatter(x, y2, s=2, c='r', marker="o", label='food')
plt.ylabel("volume")
plt.xlabel('Months')

plt.legend(loc='upper left');
plt.show()
plt.savefig('Matplotlib_save_plot.png')
```
